#### 1. Mathematical Function Plotting (Matplotlib)

Matplotlib is the core library used to visualize mathematical theory, such as plotting equations like the following equation.

$$ y=x^{2} $$

**Main Functions:**
- plt.figure(): Initializes a new drawing canvas.
- plt.plot(x, y): Draws lines or markers connecting data points.
- plt.scatter(x, y): Creates a scatter plot for observing point distributions.
- plt.title(), plt.xlabel(), plt.ylabel(): Adds context through labels.
- plt.legend(): Identifies different data series in the same plot.
- plt.show(): Renders and displays the final plot.

```python
# Plotting

# Compute the X and Y coordinates for points on a Since curve

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x,y)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='Sine Wave')
plt.plot(x, y_cos, label='Cosine Wave')
plt.title('Sine and Cosine Waves')
plt.xlabel('X Axis values from 0 to 4')
plt.ylabel('Y Axis Values')
plt.legend()
plt.show()
```

```python
# Subplot
# Compute the X and Y coordinates for points on a Since and Cosine curve

x = np.arange(0, 4*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up the subplots grid that has height 2 and width 1
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, y_sin, 'b-', label='Sine Wave')
plt.title('Sine Wave')

# Second subplot, as active, and make the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'r-', label='Cosine Wave')
plt.title('Cosine Wave')

plt.tight_layout()  # Adjust subplots to fit into figure area.
plt.show()
```

#### 2. Graph Theory Visualization (NetworkX)

Graph theory involves nodes (vertices) and edges (connections). The NetworkX library is the standard for this.

**Main Functions:**

- nx.Graph(): Creates an undirected graph object.
- G.add_node() / G.add_edge(): Adds theoretical components to the graph.
- nx.draw(G): Visualizes the network structure.
- nx.shortest_path(G, source, target): Implements graph algorithms like finding the quickest route.

```python
import networkx as nx

# Create a theoretical graph structure
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A'), ('A', 'D')])

# Plot the graph
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

#### Numpy

NumPy (Numerical Python) is the foundational library for scientific computing in Python. As of 2026, it remains the industry standard for handling large, multi-dimensional arrays and matrices with high performance.

- https://numpy.org/doc/stable/#

**Core Theory**

- **The ndarray Object:** The center of NumPy is the ndarray (n-dimensional array), which stores homogeneous data (all elements must be the same type).
- **Vectorization:** This allows you to perform operations on entire arrays at once without writing explicit for loops, significantly increasing execution speed.
- **Broadcasting:** A powerful mechanism that allows NumPy to work with arrays of different shapes during arithmetic operations.
- **Performance:** NumPy arrays are stored in contiguous memory blocks, making them up to 50x faster than standard Python lists.

```python
import numpy as np

# 1. Array Creation

# Create from a list
arr = np.array([1, 2, 3])

# Create from a list
arr = np.array([1, 2, 3])
print("Array:", arr) 

# Arrays with placeholders
print("Zeros:\n", np.zeros((2, 2)))

print("Range (0 to 10 step 2):", np.arange(0, 10, 2))

print("Linspace (0 to 1, 5 parts):", np.linspace(0, 1, 5))
```

```python
# 2. Array Manipulation

arr_2d = np.array([[1, 2], [3, 4], [5, 6]])

# Reshaping
print("Reshaped (2x3):\n", arr_2d.reshape(2, 3))

# Transposing
print("Transposed:\n", arr_2d.T)

# Flattening
print("Flattened:", arr_2d.flatten())
```

```python
# 3. Mathematical Operations

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

# Vectorized Arithmetic
print("Addition:", a + b)    
print("Multiplication:", a * b)

# Aggregations
print("Sum:", np.sum(a))       
print("Mean:", np.mean(a))     
print("Max index:", np.argmax(a))
```

```python
# 4. Linear Algebra

m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

# Matrix Multiplication (Dot Product)
print("Matrix Product:\n", np.matmul(m1, m2))

# Inverse Matrix
print("Inverse of m1:\n", np.linalg.inv(m1))
```

```python
# 5. Indexing and Slicing

arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])

# Access element at row 1, column 2
print("Element at (1,2):", arr[1, 2])

# Slice rows 0 to 2 and columns 1 to 3
print("Sliced array:\n", arr[0:2, 1:3])

# Boolean indexing
print("Elements > 50:", arr[arr > 50])
```

```python
# Count migrants (excluding US natives)
df = data[data['native-country'] != 'United-States']
country_counts = df['native-country'].value_counts().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(8, 3), dpi=300)
bars = ax.bar(country_counts.index, country_counts.values)
plt.xticks(rotation=45, ha='right', fontsize=6)
plt.ylabel('Number of non-US natives')
plt.title('Migrants by Native Country')
plt.show()
```

```python
# Identify occupation with more males than females
occupation_counts = data.groupby(['occupation', 'sex']).size().unstack()
occupation_counts['Total'] = occupation_counts.sum(axis=1)
occupation_counts['Percent Male'] = (occupation_counts['Male'] / occupation_counts['Total']) * 100
occupation_counts = occupation_counts.sort_values('Percent Male', ascending=False)
occupation_counts.reset_index(inplace=True)
display(occupation_counts)
```
